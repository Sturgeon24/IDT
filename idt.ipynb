```python
import matplotlib.pyplot as plt
import numpy as np

# Define years from the present to 10 years into the future
years = np.arange(2025, 2035)

# Hypothetical data for different energy sources (in gigawatts or similar units)
wind_energy = np.linspace(200, 400, len(years))  # Steady growth
electric_vehicles = np.linspace(50, 300, len(years))  # Rapid growth
hydrogen_energy = np.linspace(20, 150, len(years))  # Emerging energy source
bio_energy = np.linspace(80, 150, len(years))  # Moderate growth
nuclear_energy = np.linspace(300, 350, len(years))  # Stable, slow growth

# Create a plot
plt.figure(figsize=(12, 6))

# Plot each energy source
plt.plot(years, wind_energy, label="Wind Energy", marker="o")
plt.plot(years, electric_vehicles, label="Electric Vehicles", marker="s")
plt.plot(years, hydrogen_energy, label="Hydrogen Energy", marker="^")
plt.plot(years, bio_energy, label="Bioenergy", marker="d")
plt.plot(years, nuclear_energy, label="Nuclear Energy", marker="x")

# Customize the plot
plt.title("Projected Growth of Sustainable Energy Sources (2025-2035)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Energy Contribution (Gigawatts)", fontsize=14)
plt.xticks(years, rotation=45)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```
